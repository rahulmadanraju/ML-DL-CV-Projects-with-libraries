```python
import numpy as np
import matplotlib.pyplot as plt

import cv2

img = cv2.imread('2.png')
plt.imshow(img)
newimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(newimg)

font = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText(newimg, text = 'Rahul!', org = (10,500), fontFace = font, fontScale = 4, color=(0,255,0), thickness=5, lineType = cv2.LINE_AA )

plt.imshow(newimg)
```
